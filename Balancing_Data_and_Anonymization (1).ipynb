```python
import pandas as pd
data = pd.read_parquet('finalData1.parquet')
```

```python
from sklearn.utils import resample


negative = data[data.Lable == 'negative']
positive = data[data.Lable == 'positive']
neutral = data[data.Lable == 'neutral']

n_samples = len(neutral)  # Number of samples in the neutral class

# Resample
negative_resampled = resample(negative, replace=False, n_samples=n_samples, random_state=123)
positive_resampled = resample(positive, replace=False, n_samples=n_samples, random_state=123)

# Combine and shuffle
data_balanced = pd.concat([negative_resampled, positive_resampled, neutral])
data_balanced = data_balanced.sample(frac=1, random_state=123).reset_index(drop=True)

print(data_balanced.shape[0])
data_balanced.head()
```

```python
data_balanced = data_balanced.drop('CorrectedText', axis=1)
```

```python
import pandas as pd
from flair.data import Sentence
from flair.models import SequenceTagger


tagger = SequenceTagger.load('flair/ner-english-ontonotes-large')

def process_text(text):

    sentence = Sentence(text)
    tagger.predict(sentence)

    # Replace entities with their group name
    for entity in sentence.get_spans('ner'):
        if entity.tag in ['CARDINAL', 'DATE', 'EVENT', 'FAC', 'GPE', 'LANGUAGE', 'LAW', 'LOC', 'MONEY', 'NORP', 'ORDINAL', 'ORG', 'PERCENT', 'PERSON', 'PRODUCT', 'QUANTITY', 'TIME', 'WORK_OF_ART']:
            text = text.replace(entity.text, entity.tag)
    return text


#data.loc[:1999, 'Text'] = data.loc[:1999, 'Text'].apply(process_text)
data_balanced.loc[:2000, 'Text1'] = data_balanced.loc[:2000, 'Text'].apply(process_text)



data_balanced.head()
```
