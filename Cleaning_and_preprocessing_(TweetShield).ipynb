```python
import re
import string
from nltk.corpus import wordnet

def cleanTweetText(text: str) -> str:
    # Remove hashtags
    removeHash = text.replace("#", " ")

    #remove _ and -
    removeUnderline =  removeHash.replace('_', ' ').replace('-', ' ')

    # Split words joined by capital letters
    splitJoinedWords = re.sub(r"(?<=[a-z])(?=[A-Z])", " ", removeUnderline)

    # Convert to lowercase
    #lowercase = splitJoinedWords.lower()

    # Remove new line characters
    removeNewLineChar = splitJoinedWords.replace('\n', " ").replace('\r', " ")

    # Remove numbers
  #  removeNumbers = re.sub(r'[0-9]+', " ", removeNewLineChar)

    # Remove punctuation
  #  removePunctuation = removeNewLineChar.translate(str.maketrans('', '', string.punctuation))

    # Remove non-English words
  #  removeNonEnglishWords = " ".join(w for w in removeNewLineChar.split() if wordnet.synsets(w))

    # Remove single letters
  #  removeSingleLetters = " ".join([word for word in removeNonEnglishWords.split() if len(word) >= 2])

    # Strip extra spaces from the beginning and end of the string
    removeOuterSpaces = removeNewLineChar.strip()

    return removeOuterSpaces

```

```python
from emot.core import emot  # Correctly import the emot class from the emot module

def replace_emojis_and_emoticons_with_meanings(text):
    emot_obj = emot()  # Create an instance of the emot class

    # Detect emojis and replace them with meanings
    emoji_result = emot_obj.emoji(text)
    if emoji_result['flag']:  # Check if any emoji was found
        for value, mean in zip(emoji_result['value'], emoji_result['mean']):
            text = text.replace(value, f"{mean}")  # Replace emoji with its meaning

    # Detect emoticons and replace them with meanings
    emoticon_result = emot_obj.emoticons(text)
    if emoticon_result['flag']:  # Check if any emoticon was found
        for value, mean in zip(emoticon_result['value'], emoticon_result['mean']):
            text = text.replace(value, f"{mean}")  # Replace emoticon with its meaning

    return text

# Example usage
text = "I love python  :) "
replaced_text = replace_emojis_and_emoticons_with_meanings(text)
print(replaced_text)
```
