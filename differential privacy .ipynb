```python
from peft import AutoPeftModelForSequenceClassification
from transformers import AutoTokenizer, BitsAndBytesConfig
import torch
import evaluate
from datasets import load_dataset
from torch.nn import CrossEntropyLoss
```

```python
bnb_config = BitsAndBytesConfig(
load_in_4bit=True,
bnb_4bit_use_double_quant=True,
bnb_4bit_quant_type="nf4",
bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
id2label = {0: "negative", 1: "neutral", 2: "positive"}
label2id = {"negative": 0, "neutral": 1, "positive": 2}
```

```python
epochs = ['/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-1363','/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-2726',
         '/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-4089','/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-5452',
         '/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-6815','/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-8178',
         '/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-9541','/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-10904',
         '/projects/privacyllm/tweet_shield_dp_trained_model/checkpoint-12267']
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import auc
```

```python
import csv
import pandas as pd
```

```python
#import numpy as np
epochNum = 0
for epoch in epochs:
    
    epochNum+=1
        
    print("epoch number: ", epochNum)
    
    
    device = "cuda"
    model = AutoPeftModelForSequenceClassification.from_pretrained(
    epoch,
    num_labels=3,
    quantization_config=bnb_config
    )
    
    model = model.to(device)
    #print(model.eval())
    
    torchE = str(epoch) + '/score.original_module.pt'
    score_weights = torch.load(torchE,map_location="cpu")
    model.score.original_module.load_state_dict(score_weights)
    
    tokenizer = AutoTokenizer.from_pretrained(epoch)
    
    #model.device
    
    orig_dataset = load_dataset("parquet", data_files="./data_balanced.parquet")
    #orig_dataset = load_dataset("parquet", data_files="./anonymized.parquet")

    dataset = orig_dataset.rename_column("Lable", "label").rename_column("Text", "text")
    dataset = dataset["train"].train_test_split(test_size=0.1, seed=42)

    def preprocess_fn(examples):
      return tokenizer(examples["text"], truncation=True)

    toked = dataset.map(preprocess_fn, batched=True)

    def change_labels(example):
      label = example["label"]
      n_label = 0 if label == "negative" else 1 if label == "neutral" else 2
      example["label"] = n_label
      return example

    toked = toked.map(change_labels)

    accuracy = evaluate.load("accuracy")
  
    
    loss_fn = CrossEntropyLoss()
    with torch.no_grad():
        text = '"hubby and i enjoyed playing golf for only $MONEY a person. new golf course having what they call DATE madness. cool for us!! "'
        logits = model(**tokenizer(text, return_tensors="pt").to(device))["logits"]
        smax = torch.softmax(logits, dim=-1)
        loss = -torch.log(smax[0,2])
    #print(logits)
    #print(smax)
    #print(loss)
    
    
    train_losses = []
    for i,data in enumerate(toked["train"]):
        with torch.no_grad():
            logits = model(input_ids=torch.tensor([data["input_ids"]]).to(device))["logits"]
            train_losses.append(-torch.log(torch.softmax(logits,dim=-1)[0,data["label"]]))
        if i > 1000:
            break
    
    train_losses = torch.stack(train_losses)
    
    #train_losses.mean()
    
    test_losses = []
    for i,data in enumerate(toked["test"]):
        with torch.no_grad():
            logits = model(input_ids=torch.tensor([data["input_ids"]]).to(device))["logits"]
            test_losses.append(-torch.log(torch.softmax(logits,dim=-1)[0,data["label"]]))
        if i > 1000:
            break
    test_losses = torch.stack(test_losses)
    test_losses.mean()
    
    print(train_losses.shape)
    print(test_losses.shape)
    train_sorted, indices = torch.sort(train_losses)

    def TPR_FPR(index):
        FPR = index/train_sorted.shape[0]
        true_postive = (test_losses >= train_sorted[-index]).sum()
        TPR = true_postive/test_losses.shape[0]
        return TPR.item(), FPR

    TPR_FPR_list = []

    for i in range(1,train_sorted.shape[0]):
        TPR_FPR_list.append(TPR_FPR(i))
        #print(TPR_FPR(i))
        
    TPRs = []
    FPRs = []
    for item in TPR_FPR_list:
        tpr, fpr = item 
        TPRs.append(tpr)
        FPRs.append(fpr)


    # Calculate the AUC
    roc_auc = auc(FPRs, TPRs)

    # Plot the ROC curve
    plt.figure()
    plt.plot(FPRs, TPRs, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Characteristic of training with differential privacy')
    plt.legend(loc="lower right")
    plt.show()
    
    

    fixed_FPRs = [0.01, 0.05, 0.25]
    all_TPRs_at_fixed_FPRs = {} 
    
    # Interpolate TPRs at the fixed FPRs
    all_TPRs_at_fixed_FPRs[epochNum-1] = {fpr: np.interp(fpr, FPRs, TPRs) for fpr in fixed_FPRs}

    print("TPR at fixed FPR=[0.01,0.05,0.25] : " , all_TPRs_at_fixed_FPRs[epochNum-1] )
    

    csv_file = f'DP_{epochNum}.csv' 

    with open(csv_file, mode='w', newline='') as file:
        writer = csv.writer(file)

        writer.writerow(['TPR', 'FPR', 'Epoch'])

        for tpr, fpr in TPR_FPR_list:
            writer.writerow([tpr, fpr, epochNum])
```
